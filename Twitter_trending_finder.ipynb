```python
import os
import sys
import json
import tweepy
import geocoder
import pandas as pd
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
log = pd.read_csv('Login.csv')
consumerKey = log['key'][0]
consumerSecret = log['key'][1]
accesToken = log['key'][2]
accesTokenSecret = log['key'][3]
authenticate = tweepy.OAuthHandler(consumerKey, consumerSecret)
authenticate.set_access_token(accesToken, accesTokenSecret)
api = tweepy.API(authenticate, wait_on_rate_limit = True)
```

```python
def get_trends(api, loc):
    g = geocoder.osm(loc)
    closest_loc = api.trends_closest(g.lat, g.lng)
    trends = api.trends_place(closest_loc[0]["woeid"])
    return trends[0]["trends"]
```

```python
if __name__ == "__main__":
    loc = "Mexico"
    trends = get_trends(api, loc)
    print(json.dumps(trends, indent=4))
```

```python
def extract_hashtags(trends):
    hashtags = [trend["name"] for trend in trends if "#" in trend["name"]]
    return hashtags
```

```python
if __name__ == "__main__":
    hashtags = extract_hashtags(trends)
    for hashtag in hashtags:
        print(hashtag)
```
